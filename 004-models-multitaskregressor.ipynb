```python
import utils
import numpy as np
from deepchem import deepchem as dc
from deepchem.models import MultitaskRegressor
```

```python
utils.set_seed(50, False, True)
```

```python
data = utils.load_data('ABC_smiles_data.csv', 600, True)
```

```python
featurizer = dc.feat.CircularFingerprint(size=1024)
train_data, valid_data, test_data, transformer = utils.receptor_data(data, 'BCRP', featurizer)
```

```python
metric = dc.metrics.Metric(dc.metrics.rms_score)

model = MultitaskRegressor(
    1,
    1024,
    [1000, 500, 250]
)

train_model = utils.fit_best_pt(
    model,
    train_data,
    valid_data,
    [metric],
    [transformer],
    patience=10,
    model_name="MTR"
)
```

```python
utils.plot_predictions(model, train_data, test_data, transformer)
```

```python
utils.eval(model, test_data, [transformer])
```

```python
utils.plot_validation(train_model, metric)
```

```python
MTR06 = MultitaskRegressor(1, 1024, [1000, 500])
MTR06.restore(model_dir='models/MTR06.ckpt/')
```

```python
utils.plot_predictions(MTR06, train_data, test_data, transformer)
utils.eval(MTR06, test_data, [transformer])
```
